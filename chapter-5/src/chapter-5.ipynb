## ノック６４：カラーヒストグラムを可視化してみよう

```python
hist_b = cv2.calcHist([img], channels=[0], mask=None, histSize=[256], ranges=[0,256])
print(hist_b.shape)
hist_b[:5]
```

```python
hist_g = cv2.calcHist([img], channels=[1], mask=None, histSize=[256], ranges=[0,256])
hist_r = cv2.calcHist([img], channels=[2], mask=None, histSize=[256], ranges=[0,256])

import matplotlib.pyplot as plt
plt.plot(hist_r, color="r", label="Red")
plt.plot(hist_g, color="g", label="Green")
plt.plot(hist_b, color="b", label="Blue")
plt.legend()
plt.show()
```

```python
img_resized = cv2.resize(img, (300,500))
print(img_resized.shape)
cv2.imshow("img_resized.jpg", img_resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_resized_2 = cv2.resize(img, None, fx=1.5, fy=1.5)
print(img_resized_2.shape)
cv2.imshow("img_resized_2.jpg", img_resized_2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_resized_2 = cv2.resize(img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_NEAREST)
print(img_resized_2.shape)
cv2.imshow("img_resized_2.jpg", img_resized_2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## ノック６７：画像を回転させてみよう

```python
cv2.imshow("img_resized.jpg", img_resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_resized.shape)
```

```python
img_rotated = cv2.rotate(img_resized, cv2.ROTATE_90_CLOCKWISE)
cv2.imshow("img_rotated.jpg", img_rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_rotated.shape)
```

```python
height, width = img_resized.shape[:2]
center = (int(width/2), int(height/2))

rot = cv2.getRotationMatrix2D(center, 45, 1)
img_rotated = cv2.warpAffine(img_resized, rot, (width, height))
cv2.imshow("img_rotated.jpg", img_rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_rotated.shape)
```

```python
img_reverse = cv2.flip(img_resized, 0)
cv2.imshow("img_reverse.jpg", img_reverse)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_reverse.shape)
```

```python
img_reverse = cv2.flip(img_resized, 1)
cv2.imshow("img_reverse.jpg", img_reverse)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_reverse.shape)
```

```python
img_reverse = img_resized[:, ::-1, :]
cv2.imshow("img_reverse.jpg", img_reverse)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_reverse.shape)
```

## ノック６８：画像処理をしてみよう

```python
img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
cv2.imshow("img_gray.jpg", img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_gray.shape)
```

```python
th, img_th = cv2.threshold(img_gray, 60, 256, cv2.THRESH_BINARY)
cv2.imshow("img_th.jpg", img_th)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_th.shape)
```

```python
img_smoothed = cv2.blur(img_resized, (8,8))
cv2.imshow("img_smoothed.jpg", img_smoothed)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_smoothed.shape)
```

```python
img_smoothed = cv2.blur(img_resized, (20,20))
cv2.imshow("img_smoothed.jpg", img_smoothed)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img_smoothed.shape)
```

## ノック６９：画像にテキストや線を描画してみよう

```python
text = '9'
xy = (200, 100)
font = cv2.FONT_HERSHEY_COMPLEX
font_scale = 2
color = (0,0,255)
thickness = 2

img_text = cv2.putText(img_resized.copy(), text, xy, font, font_scale, color, thickness)
cv2.imshow("img_text.jpg", img_text)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
x0, y0 = 200, 70
x1, y1 = 350, 330
color = (0,0,255)
thickness = 3

img_rect = cv2.rectangle(img_resized.copy(), (x0,y0), (x1, y1), color, thickness)
cv2.imshow("img_rect.jpg", img_rect)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
